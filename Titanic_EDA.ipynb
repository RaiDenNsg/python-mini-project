Part 1 : Univariate Analysis

```python
sns.histplot(df['age'], kde = True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.boxplot(x=df['age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
# Univariate Analysis - Categorical Feature
```

```python
sns.countplot(x='sex', data = df)
plt.title('Gender Count')
plt.show()
```

```python
sns.countplot(x='survived', data=df)
plt.title('Survival Count(0 = No, 1 = Yes)')
plt.show()
```

```python
sns.countplot(x='class', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

Bivariate Analysis

```python
sns.boxplot(x='survived', y='age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.boxplot(x='survived', y='fare', data=df)
plt.title('Fare vs Survival')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Gender vs Survival')
plt.show()
```

```python
sns.countplot(x='class', hue='survived',data=df)
plt.title('Class vs Survival')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=df)
plt.title('Age vs Fare')
plt.show()
```

Multivariate Analysis

```python
corr = df.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.catplot(
    x='class',
    hue='sex',
    col='survived',
    data=df,
    kind='count',
)
plt.show()
```

```python
sns.pairplot(
    df,
    hue='survived',
    vars=['age', 'fare']
)
plt.show()
```

Final Conclusion
- Females survived more than males.
- Higher-class passengers had better chances of survival.
- Higher fare --> higher survival probability
- Children are more likely to survive.
- Third-class males passengers were the most vulnerable group.
